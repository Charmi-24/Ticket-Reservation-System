```python
# Defining a function for verifying the login credentials
def login_Verif(username, password):
    with sqlite3.connect("TRS.db") as db:
        cursor = db.cursor()
    query1 = "SELECT count(*) FROM user WHERE username= '{}' AND password= '{}';"
    cred = cursor.execute(query1.format(username, password))
    for data in cred:
        if data[0] == 1:
            print('Login Successful') 
        else:
            print('User does not exist')
    db.commit()
    db.close()
```

```python
# Defining customer class
class Customer:
    
    def __init__(self, username): 
        self.customername = username
        
# Defining class for booking the tickets
class Booking(Customer): 

    # To display the schedule of the trains
    def getTrainsInfo(self): 
        print("Schedule of the trains")
        print("Train_No   |  Train_Name   |  Journey  |  Available_Seats  |  Charges")
        print('-'*80)
        
        with sqlite3.connect("TRS.db") as db:
            cursor = db.cursor()
        trainQuery = "SELECT * FROM train;"
        trainData = cursor.execute(trainQuery)
        for data in trainData:
            listData = list(data)
            print(listData[0],'   |  ',listData[1],'   |  ',listData[2],'   |  ',listData[3],'   |  ',listData[4])
        
        db.commit()
        db.close()
        
    # To book the tickets as per the schedule    
    def bookTickets(self, journey, seatCount, train_no):
        user = self.customername
        with sqlite3.connect("TRS.db") as db:
            cursor = db.cursor()
        getQuery = "SELECT * FROM train WHERE trainno = '{}';".format(train_no.get())
        a = cursor.execute(getQuery)
        data = a.fetchall()
        charge = 0
        for i in data:
            charge = i[4]
        finalCharge = charge * seatCount.get()
        insertQuery = "INSERT INTO bookings (trainno, journey, seats, charges, user) VALUES ({}, '{}', {}, {},'{}');".format(train_no.get(), journey.get(), seatCount.get(), finalCharge, user)
        trainData = cursor.execute(insertQuery)
        print("Inserted")
        db.commit()
        db.close()
        
        journey_input.delete(0, END)
        seatCount_input.delete(0, END)
        train_no_input.delete(0, END)
    
    # To view the tickets booked
    def viewtickets(self): 
        with sqlite3.connect("TRS.db") as db:
            cursor = db.cursor()
        viewQuery = "SELECT * FROM bookings;"
        a = cursor.execute(viewQuery)
        data = a.fetchall()
        print("Train_No   |  Journey   |  Seats Count  |  Charges  |  User")
        for i in data:
            listData = list(i)
            print(listData[0],'   |  ',listData[1],'   |  ',listData[2],'   |  ',listData[3],'   |  ',listData[4])
        db.commit()
        db.close()
```
